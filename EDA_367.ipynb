```python
import pandas as pd
import numpy as np

excel_file = pd.ExcelFile('/Users/adamorencia/Desktop/BDI367/fbc_data_2026.xlsx')
df = pd.read_excel('/Users/adamorencia/Desktop/BDI367/fbc_data_2026.xlsx', sheet_name='County')

print("Dataset Shape:", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
print("\nMissing values:")
print(df.isnull().sum())
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Check monthly vs yearly columns
print("Sample data with first few rows:")
print(df[['Housing', 'Housing.1', 'Food', 'Food.1', 'Total', 'Total.1']].head(10))

print("\n\nRatio check (multiply monthly by 12):")
for col in ['Housing', 'Food', 'Transportation', 'Healthcare', 'Taxes']:
    print(f"\n{col}:")
    print(f"  Monthly (avg): ${df[col].mean():.2f}")
    print(f"  Yearly .1 (avg): ${df[col + '.1'].mean():.2f}")
    print(f"  Monthly × 12: ${df[col].mean() * 12:.2f}")
    print(f"  Match? {abs(df[col].mean() * 12 - df[col + '.1'].mean()) < 1}")
```

```python
print("Total columns:", len(df.columns))
print("Total rows:", len(df))
```

```python
print("\nAffordability Ratio Analysis")
print(df['affordability_ratio'].describe())
print(f"\nMedian: {df['affordability_ratio'].median():.3f}")
print(f"Ratio > 1.5: {(df['affordability_ratio'] > 1.5).sum()} ({(df['affordability_ratio'] > 1.5).sum()/len(df)*100:.1f}%)")
print(f"Ratio < 0.8: {(df['affordability_ratio'] < 0.8).sum()} ({(df['affordability_ratio'] < 0.8).sum()/len(df)*100:.1f}%)")
```

```python
print("\nFamily Type Analysis")
print("\nFamily Types:")
print(df['Family'].value_counts())

family_aff = df.groupby('Family')[['affordability_ratio', 'median_family_income']].agg(['mean', 'median', 'min', 'max']).round(3)
print("\nAffordability by Family Type:")
print(family_aff)
```

```python
print("\nState-Level Geographic Analysis")

state_analysis = df.groupby('State abv.')[['affordability_ratio', 'median_family_income']].agg(['mean', 'std', 'count']).round(3)
state_analysis.columns = ['Aff_Ratio_Mean', 'Aff_Ratio_Std', 'Aff_Ratio_Count', 'Income_Mean', 'Income_Std', 'Income_Count']

print("\nLeast Affordable States:")
print(state_analysis.sort_values('Aff_Ratio_Mean', ascending=False).head(10)[['Aff_Ratio_Mean', 'Income_Mean']])

print("\nMost Affordable States:")
print(state_analysis.sort_values('Aff_Ratio_Mean', ascending=True).head(10)[['Aff_Ratio_Mean', 'Income_Mean']])
```

```python
print("\nExpense Category Analysis")

expense_cols = ['Housing', 'Food', 'Transportation', 'Healthcare', 'Other Necessities ', 'Childcare', 'Taxes']
total_col = 'Total'

print("\nAverage Expenses by Category:")
expense_summary = df[expense_cols].describe().loc[['mean', '50%', 'min', 'max']].T
expense_summary.columns = ['Mean', 'Median', 'Min', 'Max']
print(expense_summary.round(0))

print("\nExpense Distribution (% of Total):")
expense_pct = (df[expense_cols].sum() / df[total_col].sum() * 100).sort_values(ascending=False)
print(expense_pct.round(2))

print("\nCorrelation with Affordability Ratio:")
correlations = df[expense_cols + ['affordability_ratio']].corr()['affordability_ratio'].drop('affordability_ratio').sort_values(ascending=False)
print(correlations.round(3))
```
